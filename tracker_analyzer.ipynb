```python
import json
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from mpl_toolkits.mplot3d import Axes3D
```

```python
with open("C:\\Users\\shrim\\tracker\\data_log2.txt", "r") as f:
    data_log = json.load(f)
data_array = np.array(data_log)
if data_array.ndim != 2 or data_array.shape[1] != 2:
    raise ValueError("Invalid data_log format: Expected list of 2-item lists.")
time_col = np.arange(0, len(data_array) * 0.1, 0.1)
data_with_time = np.column_stack((time_col, data_array))
```

```python
plt.figure(figsize=(75, 10))
plt.plot(time_col, data_array[:, 0], label="Potentiometer 1", marker="o")
plt.plot(time_col, data_array[:, 1], label="Potentiometer 2", marker="s")
plt.xlabel("Time (seconds)")
plt.ylabel("Potentiometer Value")
plt.title("Potentiometer Readings Over Time")
plt.legend()
plt.grid()
plt.show()
```
