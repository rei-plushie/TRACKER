```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load data from data_log.txt
with open("C:\\Users\\shrim\\tracker\\data_log.txt", "r") as f:
    data_log = json.load(f)  # Read as list of lists

# Convert to NumPy array
data_array = np.array(data_log)

# Ensure data is in correct shape (N, 2)
if data_array.ndim != 2 or data_array.shape[1] != 2:
    raise ValueError("Invalid data_log format: Expected list of 2-item lists.")

# Generate time column (0, 0.25, 0.50, ...)
time_col = np.arange(0, len(data_array) * 0.25, 0.25)

# Combine time with data
data_with_time = np.column_stack((time_col, data_array))
```

```python
# Plot data
plt.figure(figsize=(75, 10))
plt.plot(time_col, data_array[:, 0], label="Potentiometer 1", marker="o")
plt.plot(time_col, data_array[:, 1], label="Potentiometer 2", marker="s")

# Formatting
plt.xlabel("Time (seconds)")
plt.ylabel("Potentiometer Value")
plt.title("Potentiometer Readings Over Time")
plt.legend()
plt.grid()

# Show plot
plt.show()
```
